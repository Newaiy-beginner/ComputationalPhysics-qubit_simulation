```python
import math
import cmath
import numpy as np
import matplotlib.pyplot as plt

#input E0
E0 = 2.0

hbar = 1.0
A = 3.0    #Bx magnitude

E1 = E0
E2 = E0

T0 = math.pi/(math.sqrt(E0**2.0 + A**2.0))
T1 = math.pi/A

omega0 = (E1 - E2)/hbar   #equiv to Bz
theta = math.atan(omega0/(2.0*A/hbar))
print(theta)
c10 = 1.0
c20 = math.sqrt(1.0 - abs(c10)**2.0)
'''
hbar*Bz equiv. to E1 - E2
hbar*Bx/2 equiv. to A(offdiagonal term)
'''
mid = 0.0
def V1(x):
    if x < mid:
        return 0.0
    elif x == mid:
        return A/2.0
    else:
        return A

def V2(x):
    err = math.erf(0.45*(x - mid))+1.0
    return complex(A*err/2.0)

def g1(x, y1, y2):
    x = x.real
    V = V1(x)
    if (x >= 0.0 and x <= T0) or (x >= (T0 + 0.5*T1) and x <= (2.0*T0 + 0.5*T1)):
        return ((-1j)/hbar)*(E1*y1 + V*y2)
    else:
        return ((-1j)/hbar)*(V*y2)

def gg1(x, y1, y2):
    x = x.real
    V = V2(x)
    if (x >= 0.0 and x <= T0) or (x >= (T0 + 0.5*T1) and x <= (2.0*T0 + 0.5*T1)):
        return ((-1j)/hbar)*(E1*y1 + V*y2)
    else:
        return ((-1j)/hbar)*(V*y2)

def g2(x, y1, y2):
    x = x.real
    Vbar = V1(x)
    if (x >= 0.0 and x <= T0) or (x >= (T0 + 0.5*T1) and x <= (2.0*T0 + 0.5*T1)):
        return ((-1j)/hbar)*(Vbar*y1 + E2*y2)
    else:
        return ((-1j)/hbar)*(Vbar*y1)
    
def gg2(x, y1, y2):
    x = x.real
    Vbar = V2(x)
    if (x >= 0.0 and x <= T0) or (x >= (T0 + 0.5*T1) and x <= (2.0*T0 + 0.5*T1)):
        return ((-1j)/hbar)*(Vbar*y1 + E2*y2)
    else:
        return ((-1j)/hbar)*(Vbar*y1)

def rk4(f1, f2, x0, y10, y20, x1, n):
    vx = np.zeros(n+1, complex)
    vy1 = np.zeros(n+1, complex)
    vy2 = np.zeros(n+1, complex)
    h = complex((x1 - x0)/n)
    vx[0] = x = x0
    vy1[0] = y1 = y10
    vy2[0] = y2 = y20
    for i in range(1, n+1):
        c1g1 = h * f1(x, y1, y2)
        c2g1 = h * f2(x, y1, y2)
        c1g2 = h * f1(x + 0.5*h, y1 + 0.5*c1g1, y2 + 0.5*c2g1)
        c2g2 = h * f2(x + 0.5*h, y1 + 0.5*c1g1, y2 + 0.5*c2g1)
        c1g3 = h * f1(x + 0.5*h, y1 + 0.5*c1g2, y2 + 0.5*c2g2)
        c2g3 = h * f2(x + 0.5*h, y1 + 0.5*c1g2, y2 + 0.5*c2g2)
        c1g4 = h * f1(x + h, y1 + c1g3, y2 + c2g3)
        c2g4 = h * f2(x + h, y1 + c1g3, y2 + c2g3)
        vx[i] = x = x0 + i*h
        vy1[i] = y1 = y1 + (c1g1 + 2*c1g2 + 2*c1g3 + c1g4)/6
        vy2[i] = y2 = y2 + (c2g1 + 2*c2g2 + 2*c2g3 + c2g4)/6
    return vx.real, vy1, vy2

t, c1, c2 = rk4(g1, g2, 0.0, c10, c20, 2.0, 200)
tprime, cc1, cc2 = rk4(gg1, gg2, 0.0, c10, c20, 12.0, 2000)

C1 = [abs(c1[i])**2.0 for i in range(len(c1))]
C2 = [abs(c2[i])**2.0 for i in range(len(c2))]
Poten1 = [V1(t[i])/A for i in range(len(t))]
Poten2 = [V2(t[i])/A for i in range(len(t))]

fig = plt.figure(figsize = (11, 7))
plt.grid(True)
# plt.plot(t, Poten1, 'g-', linewidth=3)
# plt.plot(t, Poten2, 'g-', linewidth=3)
plt.plot(t, C1, label="$|C_1|^2$")
plt.plot(t, C2, label="$|C_2|^2$")
# plt.plot(t, CC1, 'r-.', label="$|C_1^{\prime}|^2$")
plt.xlabel("Time")
plt.title("A="+str(A))
plt.legend()

```
