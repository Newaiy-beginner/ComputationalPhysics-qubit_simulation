```python
import math
import cmath
import numpy as np
import matplotlib.pyplot as plt

hbar = 1.0
A = 1.0    #Bx magnitude
E0 = 0.5
E1 = E0
E2 = (-1)*E0
omega0 = (E1 - E2)/hbar   #equiv to Bz
theta = math.atan(omega0/(2.0*A/hbar))
print(theta)
c10 = 1.0
c20 = math.sqrt(1.0 - abs(c10)**2.0)
'''
hbar*Bz equiv. to E1 - E2
hbar*Bx/2 equiv. to A(offdiagonal term)
'''

def g1(x, y1, y2):
    V = A
    return ((-1j)/hbar)*(E1*y1 + V*y2)
def g2(x, y1, y2):
    Vbar = A
    return ((-1j)/hbar)*(Vbar*y1 + E2*y2)
def rk4(f1, f2, x0, y10, y20, x1, n):
    vx = np.zeros(n+1, complex)
    vy1 = np.zeros(n+1, complex)
    vy2 = np.zeros(n+1, complex)
    h = complex((x1 - x0)/n)
    vx[0] = x = x0
    vy1[0] = y1 = y10
    vy2[0] = y2 = y20
    for i in range(1, n+1):
        c1g1 = h * f1(x, y1, y2)
        c2g1 = h * f2(x, y1, y2)
        c1g2 = h * f1(x + 0.5*h, y1 + 0.5*c1g1, y2 + 0.5*c2g1)
        c2g2 = h * f2(x + 0.5*h, y1 + 0.5*c1g1, y2 + 0.5*c2g1)
        c1g3 = h * f1(x + 0.5*h, y1 + 0.5*c1g2, y2 + 0.5*c2g2)
        c2g3 = h * f2(x + 0.5*h, y1 + 0.5*c1g2, y2 + 0.5*c2g2)
        c1g4 = h * f1(x + h, y1 + c1g3, y2 + c2g3)
        c2g4 = h * f2(x + h, y1 + c1g3, y2 + c2g3)
        vx[i] = x = x0 + i*h
        vy1[i] = y1 = y1 + (c1g1 + 2*c1g2 + 2*c1g3 + c1g4)/6
        vy2[i] = y2 = y2 + (c2g1 + 2*c2g2 + 2*c2g3 + c2g4)/6
    return vx.real, vy1, vy2

t, c1, c2 = rk4(g1, g2, 0, c10, c20, 10.0, 7000)
C1 = [abs(c1[i])**2.0 for i in range(len(c1))]
C2 = [abs(c2[i])**2.0 for i in range(len(c2))]

stopt = int(0)
M = max(C2)
for i in range(len(t)):
    if abs(C1[i] - M) < 0.0001:
        stopt = i
        break
print(C1[stopt], '\t', (A**2.0)/(E0**2.0 + A**2.0))
trans = complex(0, 0)
for i in range(len(c2)):
    if C2[i] == M:
        trans = c2[i]
        break
print("trans", abs(trans)**2.0)
print("Comparison phi")
print("trans_phi =", math.atan(trans.imag/trans.real))
c1p = c1[stopt]
print("origi_phi =", math.atan(c1p.imag/c1p.real))
print("rela =", math.atan(trans.imag/trans.real)-math.atan(c1p.imag/c1p.real))

c1next = c1p*cmath.exp(1j*(math.atan(trans.imag/trans.real)-math.atan(c1p.imag/c1p.real))*2.0)
c2next = math.sqrt(1.0 - abs(c1next)**2.0)

t1, c11, c21 = rk4(g1, g2, 0, c1next, c2next, 10.0, 7000)
fig = plt.figure(figsize = (11, 7))
ax = fig.add_subplot(111)
ax.grid(True)
ax.plot(t1, c11.real, 'r-')
ax.plot(t1, c11.imag, 'r-.')
C11 = [abs(c11[i])**2.0 for i in range(len(c11))]
anal = complex(math.sqrt(A**2-E0**2)/A, (E0**2)/(A*math.sqrt(A**2+E0**2)))
anal = anal*cmath.exp(1j*(math.atan(trans.imag/trans.real)-math.atan(c1p.imag/c1p.real))*2.0)
ax.plot(t1, [anal.real for i in range(len(t1))])
ax.plot(t1, [anal.imag for i in range(len(t1))])
ax.plot(t1, C11)
ax.plot(t1, c21.real, 'g-')
ax.plot(t1, c21.imag, 'g-.', zorder = 5)
ax.axvline(2.0*math.pi/math.sqrt(A**2 + E0**2.0))


'''
def perio(arr):
    tolerance = 0.000001
    for i in range(1, len(arr)):
        if abs(arr[i]-arr[0]) < tolerance:
            return i
    return -1

fig = plt.figure(figsize = (11, 7))
ax = fig.add_subplot(111)
ax.grid(True)
ax.plot(t, [(A**2.0)/(E0**2.0 + A**2.0) for i in range(len(t))], 'c-')

phi = math.atan(-math.sqrt(A**4.0 - E0**4.0)/E0**2.0)
print("phi=", phi)
print(2.0*math.pi/E0)
c1 = cmath.exp(1j*(phi + 2*math.pi))*c1
tau = (hbar/math.sqrt(E0**2.0 + A**2.0))*math.acos(math.sqrt(1.0 - (E0/A)**2.0))
T = 2.0*math.pi*hbar/math.sqrt(E0**2.0 + A**2.0)
ax.axvline(tau)
ax.axvline(tau + T)

ax.plot(t, c1.real, 'r-')
ax.plot(t, c1.imag, 'r-.')
ax.plot(t, C1)
ax.plot(t, c2.real, 'g-')
ax.plot(t, c2.imag, 'g-.', zorder = 5)






d = (math.cos(theta))**2.0 - 0.5
print(d)
r = math.sqrt(0.25 - d**2.0)
print(r)
l = math.sin(theta)**2.0 * math.tan(theta)
print(l)
print(l/r)
phi = math.acos(l/r)
phi /= -2.0
print("phase:", phi)
print((cmath.exp(1j*phi)).imag)
print("T estimate:", 2.0*math.pi/math.sqrt((2*A)**2.0 + (omega0)**2.0))    # T = sqrt(Bx^2 + Bz^2) = sqrt((2A)^2 + (E1 - E2)^2)
'''
```
